```python
import pandas as pd

def confronta_statistiche(dataset, continente1, continente2):
    
    df = pd.read_csv(dataset)
    
    # Filtro i dati per continente
    africa = df[df['continent'] == 'Africa']
    asia = df[df['continent'] == 'Asia']
    europe = df[df['continent'] == 'Europe']
    namerica = df[df['continent'] == 'North America']
    oceania = df[df['continent'] == 'Oceania']
    samerica = df[df['continent'] == 'South America']
    
    # Aggiungo le locazioni senza indicazione di continente al totale mondiale (un giorno intero per scoprire come fare...)
    mondo = df[df['continent'].isna()]
    
    # Calcolo i descrittori statistici per i casi totali
    statistiche_continente1 = calcola_statistiche(continente1, df, mondo)
    statistiche_continente2 = calcola_statistiche(continente2, df, mondo)
    
    # Restituisce i risultati
    return statistiche_continente1, statistiche_continente2

def calcola_statistiche(continente, df, mondo):
    continente_df = df[df['continent'] == continente]
    totale_casi_mondo = mondo['total_cases'].sum()
    totale_casi_continente = continente_df['total_cases'].sum()
    
    minimo = continente_df['total_cases'].min()
    massimo = continente_df['total_cases'].max()
    media = continente_df['total_cases'].mean()
    percentuale_mondiale = (totale_casi_continente / totale_casi_mondo) * 100
    
    return {
        'minimo': minimo,
        'massimo': massimo,
        'media': media,
        'percentuale_mondiale': percentuale_mondiale
    }

# Utilizzo della funzione a partire dal file su github (un altro giorno per scoprire come farlo...)
dataset = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv" 
continente1 = input("Scegli il primo continente tra Africa, Asia, Europe, North America. Oceania e South America: ")
continente2 = input("Scegli il secondo continente tra Africa, Asia, Europe, North America. Oceania e South America: ")

statistiche_continente1, statistiche_continente2 = confronta_statistiche(dataset, continente1, continente2)

print(f"Statistiche per il continente {continente1}:")
print(statistiche_continente1)
print(f"Statistiche per il continente {continente2}:")
print(statistiche_continente2)
```

```python
import pandas as pd

def confronta_statistiche(dataset, continente1, continente2):
    
    df = pd.read_csv(dataset)
    
    filtro_continente1 = df['continent'] == continente1
    filtro_continente2 = df['continent'] == continente2
    
    # Seleziono i dati relativi ai due continenti specificati
    continente1_df = df[filtro_continente1]
    continente2_df = df[filtro_continente2]
    
    # Aggiungo le locazioni senza indicazione di continente al totale mondiale
    mondo_df = df[df['continent'].isna()]
    
    # Calcolo il totale delle vaccinazioni per ogni continente
    totale_vaccinazioni_continente1 = continente1_df['total_vaccinations'].sum()
    totale_vaccinazioni_continente2 = continente2_df['total_vaccinations'].sum()
    
    # Calcolo i descrittori statistici per i casi totali
    statistiche_continente1 = calcola_statistiche(continente1_df)
    statistiche_continente2 = calcola_statistiche(continente2_df)
    
    # Calcolo la percentuale rispetto al numero di vaccinazioni totali per ogni continente
    percentuale_continente1 = (statistiche_continente1['media'] / totale_vaccinazioni_continente1) * 100
    percentuale_continente2 = (statistiche_continente2['media'] / totale_vaccinazioni_continente2) * 100
    
    # Aggiungo le percentuali alle statistiche
    statistiche_continente1['percentuale_vaccinazioni'] = percentuale_continente1
    statistiche_continente2['percentuale_vaccinazioni'] = percentuale_continente2
    
    # Restituisce i risultati
    return statistiche_continente1, statistiche_continente2

def calcola_statistiche(continente_df):
    minimo = continente_df['total_cases'].min()
    massimo = continente_df['total_cases'].max()
    media = continente_df['total_cases'].mean()
    
    return {
        'minimo': minimo,
        'massimo': massimo,
        'media': media
    }

# Utilizzo della funzione
dataset = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
continente1 = "Europe"
continente2 = "Asia"

statistiche_continente1, statistiche_continente2 = confronta_statistiche(dataset, continente1, continente2)

print(f"Statistiche per il continente {continente1}:")
print(statistiche_continente1)
print("\n")
print(f"Statistiche per il continente {continente2}:")
print(statistiche_continente2)
```
